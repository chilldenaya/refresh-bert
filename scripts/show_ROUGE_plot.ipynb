```python
import matplotlib.pyplot as plt
import json

file_path = "/home/jupyter-23521027/refresh-bert/training/ROUGE.json"
# Read the JSON file and load data into a dictionary
with open(file_path, "r") as file:
    data = json.load(file)
    
# Training and Validation data
training_data = data["training_rouge_scores"]  
validation_data = data["validation_rouge_scores"]  

# Extract Rouge scores for training and validation
metrics = ['rouge_1_f_score', 'rouge_2_f_score', 'rouge_l_f_score']

training_scores = {metric: [data[metric] for data in training_data] for metric in metrics}
validation_scores = {metric: [data[metric] for data in validation_data] for metric in metrics}

# Epochs
epochs = range(1, len(training_data) + 1)

# Plotting
plt.figure(figsize=(36, 12))

# Rouge-1 F-score
plt.subplot(131)
plt.plot(epochs, training_scores['rouge_1_f_score'], label='Training', marker='o')
plt.plot(epochs, validation_scores['rouge_1_f_score'], label='Validation', marker='o')
plt.xlabel('Epochs')
plt.ylabel('Rouge-1 F-score')
plt.title('Rouge-1 F-score')
plt.legend()
plt.grid(True)
plt.xticks(epochs)

# Rouge-2 F-score
plt.subplot(132)
plt.plot(epochs, training_scores['rouge_2_f_score'], label='Training', marker='o')
plt.plot(epochs, validation_scores['rouge_2_f_score'], label='Validation', marker='o')
plt.xlabel('Epochs')
plt.ylabel('Rouge-2 F-score')
plt.title('Rouge-2 F-score')
plt.legend()
plt.grid(True)
plt.xticks(epochs)

# Rouge-L F-score
plt.subplot(133)
plt.plot(epochs, training_scores['rouge_l_f_score'], label='Training', marker='o')
plt.plot(epochs, validation_scores['rouge_l_f_score'], label='Validation', marker='o')
plt.xlabel('Epochs')
plt.ylabel('Rouge-L F-score')
plt.title('Rouge-L F-score')
plt.legend()
plt.grid(True)
plt.xticks(epochs)

plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd

def get_tokens(sentence):
    return sentence.split()

def process_folder(folder_path, folder_name):
    file_data = []
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r') as file:
            content = file.read()
            sentences = content.splitlines()
            tokens_list = [get_tokens(sentence) for sentence in sentences if sentence.strip()]
            token_count = sum(len(tokens) for tokens in tokens_list)
            token_count
            file_data.append({
                'File': filename,
                'Token Count': token_count,
                'Folder': folder_name,
                'Sentences': sentences,
                'Tokens': tokens_list
            })
    return file_data


folder_gold_path = "/home/jupyter-23521027/refresh-bert/data/Baseline-Gold-Models/gold-liputan6-test-orgcase"
folder_A_path = "/home/jupyter-23521027/refresh-bert/training_500_cnn_2sum/model.ckpt.epoch-11.test-summary-topranked"
folder_B_path = "/home/jupyter-23521027/refresh-bert/training_500_sbert_2sum/model.ckpt.epoch-3.test-summary-topranked"

data_gold = process_folder(folder_gold_path, 'gold')
data_A = process_folder(folder_A_path, 'A')
data_B = process_folder(folder_B_path, 'B')

# Create dataframes for Folder A and Folder B
df_gold = pd.DataFrame(data_gold)
df_A = pd.DataFrame(data_A)
df_B = pd.DataFrame(data_B)

# Concatenate both dataframes to get the final comparison dataframe
df_combined = pd.concat([df_gold, df_A, df_B], ignore_index=True)

# Group by the 'Folder' column and calculate the mean for each group
summary_by_folder = df_combined.groupby('Folder')['Token Count'].agg(['mean', 'median'])

print(summary_by_folder)
```

```python
import os
import json
import pandas as pd

folder_path = '/home/jupyter-23521027/refresh-bert/data/liputan6_data/canonical/test' 

file_list = os.listdir(folder_path)
json_data_list = []
file_list = os.listdir(folder_path)

for file_name in file_list:
    file_path = os.path.join(folder_path, file_name) 

    if os.path.isfile(file_path):
        with open(file_path, 'r') as file:
            json_data = json.load(file)
            json_data_list.append(json_data)

df = pd.DataFrame(json_data_list)
```

```python
import pandas as pd

def is_sublist(sublist, full_list):
    n = len(sublist)
    return any((sublist == full_list[i:i + n]) for i in range(len(full_list) - n + 1))

sent_indices = []  # Initialize an empty list for the Sentence Indices column

# Loop through each row in df_combined
for index, row in df_combined.iterrows():
    sentence_indices = []  # Initialize an empty list to store the indices for the current row
    for i in range(len(df)):
        if is_sublist(row['Tokens'], df['clean_article'].iloc[i]):
            sentence_indices.append(i)
    sent_indices.append(sentence_indices)

    # Print progress for every 1000 rows
    if (index + 1) % 100 == 0:
        print(f"Processed {index + 1} rows out of {len(df_combined)}")

df_combined['Sentence Indices'] = sent_indices
print(df_combined)
```
